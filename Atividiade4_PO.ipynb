### Problema 1: Onde instalar bases do SAMU

```python
prob = LpProblem("Onde instalar bases do SAMU", LpMinimize)
x1 = LpVariable('x1', cat='Binary')
x2 = LpVariable('x2', cat='Binary')
x3 = LpVariable('x3', cat='Binary')
x4 = LpVariable('x4', cat='Binary')
x5 = LpVariable('x5', cat='Binary')

bases = x1+x2+x3+x4+x5
prob += bases

norte = x4+x5
sul = x1+x2
centro = x1+x2+x3+x4
sudoeste = x2+x5
oeste = x3+x4+x5

prob += (norte>=1)
prob += (sul>=1)
prob += (centro>=1)
prob += (sudoeste>=1)
prob += (oeste>=1)

print(prob)
optimization_result = prob.solve()
print(LpStatus[prob.status])
var = [x1,x2,x3,x4,x5]
for x in var:
  print(x.name,'=',x.value())
```

```python
# Creates the 'prob' variable to contain the problem data
prob = LpProblem("PROBLEMA DO TRANSPORTE COLETIVO", LpMinimize)

# Creates a list of tuples containing all the possible routes for two types of transport
Routes = [(k, i, j) for k in range(40) for i in range(31) for j in range(31)]

# Dicionário de variáveis binárias para o uso ou não de uma determinada rota com um dos tipos dos ônibus
objective_vars = LpVariable.dicts("Uso_da_rota_pelo_transporte", (range(40) , range(31), range(31)), 0, None, LpBinary)

# Se o ônibus k foi usado ou não
Aluguel_do_onibus_K = LpVariable.dicts("Aluguel_do_onibus_K", range(40), 0, None, LpBinary)

prob += (
    lpSum([(21*objective_vars[k][i][j]*Matriz_C[k][i][j] + Aluguel_do_onibus_K[k]*Custo_Aluguel[k]/(31*31)) for (k, i, j) in Routes]),
    "Custo_total_de_transporte",
)
```
